```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
#Initializing the number of workers
workers = [1, 2, 4, 6, 8, 12]

#Intializing the array with all the time for K = 2
Tp_2 = [87.5919, 46.3124, 39.6686, 36.151, 35.9033, 38.181 ] 
Ts_2 = Tp_2[0] #Time with 1 worker(serial)
Sp_2 = [Ts_2 / i for i in Tp_2] #calculating speedup

#Speed up is calculated for all other K also

# For K = 4
Tp_4 = [1408.3351, 913.502, 70.0159, 110.5599, 489.8684, 50.3597 ] 
Ts_4 = Tp_4[0] 
Sp_4 = [Ts_4 / i for i in Tp_4] 

# For K = 5
Tp_5 = [1573.6414, 1028.9028, 920.3325, 842.3061, 58.7149 , 59.4915 ] 
Ts_5 = Tp_5[0] 
Sp_5 = [Ts_5 / i for i in Tp_5] 

# For K = 6
Tp_6 = [1758.4281, 1157.9986, 1029.9423, 923.8876, 996.7966, 65.6869  ] 
Ts_6 = Tp_6[0] 
Sp_6 = [Ts_6 / i for i in Tp_6] 



#plot for K=2
plt.plot(workers, Sp_2, color='red', label="K=2") 
plt.title('Parallel Speedup & Efficiency') 
plt.xlabel('Number of workers') 
plt.ylabel('SpeedUp Value')
plt.legend()
plt.show()
```

```python
#Plotting the speed up graph for all K values
plt.plot(workers, Sp_2, color='red', label="K=2") 
plt.plot(workers, Sp_4, color='black', label="K=4")
plt.plot(workers, Sp_5, color='green', label="K=5")
plt.plot(workers, Sp_6, color='orange', label="K=6")

plt.title('Parallel Speedup & Efficiency') 
plt.xlabel('Number of workers') 
plt.ylabel('SpeedUp Value')
plt.legend()
plt.show()
```

```python
#plotting graph for checking the execution time 
#vs number of workers for different clusters

plt.title('Execution Time for different number of workers') 
plt.xlabel('Number of workers') 
plt.ylabel('Total Time taken(s)')
plt.plot(workers, Tp_2, color='red', label="K=2") 
plt.plot(workers, Tp_4, color='black', label="K=4")
plt.plot(workers, Tp_5, color='green', label="K=5")
plt.plot(workers, Tp_6, color='orange', label="K=6")


plt.legend()
plt.show()
```
