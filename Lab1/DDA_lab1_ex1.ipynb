## Matrix Multiplication

To create a numpy matrix X of dimensions n × m, where n = 100 and
m = 20. Initialize Matrix X with random values. Then to create a numpy vector Y of dimensions m × 1.
Initialize the vector Y with values from a normal distribution using µ = 2 and σ = 0.01

```python
X=np.random.randint(10,size =(100,20)) # initializing X with random values

#Initializing vector Y with the mean and std given
Y=np.random.normal(2,0.01,(20,1)) 
print("Y vector-",Y)
```

```python
print("X matrix-",X)
print("\nShape of X:", X.shape)
```

To iteratively multiply (element-wise) each row of the matrix X with vector Y and sum the result
of each iteration in another vector C.

```python
C= np.zeros(100) #Initialize vector C
```

```python
for i in range(100):         #for each row in matrix
    sum=0
    for j in range(20):      #for each column in the row
        sum += X[i][j]* Y[j] # multiplying the required elements and adding them to the previous value to get sum
    C[i] = sum               # storing the sum of the element wise multiplication of a row
```

```python
print("Value of Vector C: ", C)
print("\nShape of vector C: ",C.shape)
```

To find mean and standard deviation of the new vector c. Here I am using the numpy functions for the same.

```python
mean = np.mean(C)     #Finds average of all the values in the vector
print("Mean of vector C:",mean)
```

```python
std = np.std(C)      # Finds the variation of the values in the vector from its mean value
print("Standard Deviation of vector C:",std)
```

Plotting the histogram of vector C using 5 bins

```python
plt.hist(C, edgecolor="red", bins=5)
plt.title("Histogram of vector C")
plt.xlabel("The range of values in vector")
plt.ylabel("The number of values ")
plt.show()
```

We observe that the histogram plot is similar to that of normal distribution. 

## GRADING

Performing tasks on the csv file given.

```python
import pandas as pd


#Reading the required csv files using pandas functionality
data_frame = pd.read_csv('Downloads/Grades.csv')

data_frame
```

* Compute the sum for all subjects for each student.

```python
#Finding sum along the subject columns
data_frame['sum'] = data_frame[['English','Science','German','Sports',' Maths']].sum(axis=1)
```

```python

data_frame
```

We observe that the column Final grade already contains the sum of all the subjects.

```python
data_frame.columns
```

* Compute the average of the point for each student(Total= 500) - 5 subjects are there

```python
#Finding average of the subjects and assigning it to new column

data_frame['average'] = data_frame[['sum']]/5
```

```python
data_frame
```

* Compute the standard deviation of point for each student.

```python
#Here we find the variation of each student from the mean of the average marks of all students. Thus we can know where
#each student is .ie, if they are above average or below average of the class.

total_avg = data_frame['average'].mean()
data_frame['standard deviation'] = data_frame['average'] - total_avg
```

```python
data_frame
```

* Plot the average points for all the students (in one figure).

```python
#Plotting the average marks of all the students
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ind = np.arange(7)
ax.set_ylabel('Average points of each student')
ax.set_xlabel('Last name of each student')
ax.set_title('Average points for all students')
ax.plot(data_frame["Last Name"],data_frame["average"])
plt.xticks(rotation=90)
plt.show()
```

* For each student assign a grade based on the rubric given


```python
conditions = [
    (data_frame['average'] < 56),
    (data_frame['average'] >= 56) & (data_frame['average'] < 60),
    (data_frame['average'] >= 60) & (data_frame['average'] < 66),
    (data_frame['average'] >= 66) & (data_frame['average'] < 70),
    (data_frame['average'] >= 70) & (data_frame['average'] < 76),
    (data_frame['average'] >= 76) & (data_frame['average'] < 80),
    (data_frame['average'] >= 80) & (data_frame['average'] < 86),
    (data_frame['average'] >= 86) & (data_frame['average'] < 90),
    (data_frame['average'] >= 90) & (data_frame['average'] < 96),
    (data_frame['average'] >= 96) & (data_frame['average'] <= 100)
]

values = ['F', 'D', 'C', 'C+','B-','B','B+','A-','A','A+']   #grade values according to the order of the conditions

data_frame['Grade'] = np.select(conditions, values) # assigning values to new column based on the conditions specified
```

```python
data_frame
```

* Plot the histogram of the final grades.

```python
plt.hist(data_frame["Grade"], edgecolor="red")
plt.title("Histogram of the final grades")
plt.xlabel("Grades")
plt.ylabel("Number of students")
plt.show()
```
