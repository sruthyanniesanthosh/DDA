```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Speed up graph

```python
#Initializing the number of workers
workers = [1, 2, 4, 6, 7]

#Intializing the array with all the time for kdd dataset
Tp_kdd = [72.8693, 54.2405, 48.0322, 48.5409, 47.7746] 
Ts_kdd = Tp_kdd[0] #Time with 1 worker(serial)
Sp_kdd = [Ts_kdd / i for i in Tp_kdd] #calculating speedup




#plot for Speedup for kdd dataset
plt.plot(workers, Sp_kdd, color='red') 
plt.title('Parallel Speedup & Efficiency') 
plt.xlabel('Number of workers') 
plt.ylabel('SpeedUp Value')

plt.show()
```

```python
#plotting graph for checking the execution time for kdd dataset


plt.title('Execution Time for different number of workers') 
plt.xlabel('Number of workers') 
plt.ylabel('Total Time taken(s)')
plt.plot(workers, Tp_kdd, color='red') 

plt.show()
```

```python
#Intializing the array with all the time for virus dataset
Tp_virus = [223.1402, 136.0188, 120.6206, 120.0451, 118.0429] 
Ts_virus = Tp_virus[0] #Time with 1 worker(serial)
Sp_virus = [Ts_virus / i for i in Tp_virus] #calculating speedup




#plot for Speedup for virus dataset
plt.plot(workers, Sp_virus, color='red') 
plt.title('Parallel Speedup & Efficiency') 
plt.xlabel('Number of workers') 
plt.ylabel('SpeedUp Value')

plt.show()
```

```python
#plotting graph for checking the execution time for virus dataset


plt.title('Execution Time for different number of workers') 
plt.xlabel('Number of workers') 
plt.ylabel('Total Time taken(s)')
plt.plot(workers, Tp_virus, color='red') 

plt.show()
```

## Convergence Graph

```python
#Reading the files storing the rmse values for kdd dataset
df_1 = pd.read_csv('kdd_results_1.csv')
df_2 = pd.read_csv('kdd_results_2.csv')
df_4 = pd.read_csv('kdd_results_4.csv')
df_6 = pd.read_csv('kdd_results_6.csv')
df_7 = pd.read_csv('kdd_results_7.csv')
```

```python
plt.title('Train Rmse Vs Epochs for different workers') 
plt.xlabel('Epochs') 
plt.ylabel('Rmse value')
plt.plot(range(0,df_1.shape[0]), df_1["Train RMSE"], color='red',label="P=1") 
plt.plot(range(0,df_2.shape[0]), df_2["Train RMSE"], color='blue',label="P=2") 
plt.plot(range(0,df_4.shape[0]), df_4["Train RMSE"], color='green',label="P=4") 
plt.plot(range(0,df_7.shape[0]), df_7["Train RMSE"], color='yellow',label="P=7") 
plt.plot(range(0,df_6.shape[0]), df_6["Train RMSE"], color='pink',label="P=6") 
plt.legend()
plt.show()
```

```python
plt.title('Test Rmse Vs Epochs for different workers') 
plt.xlabel('Epochs') 
plt.ylabel('Rmse value')
plt.plot(range(0,df_1.shape[0]), df_1["Test RMSE"], color='red',label="P=1") 
plt.plot(range(0,df_2.shape[0]), df_2["Test RMSE"], color='blue',label="P=2") 
plt.plot(range(0,df_4.shape[0]), df_4["Test RMSE"], color='green',label="P=4") 
plt.plot(range(0,df_7.shape[0]), df_7["Test RMSE"], color='yellow',label="P=7") 
plt.plot(range(0,df_6.shape[0]), df_6["Test RMSE"], color='pink',label="P=6") 
plt.legend()
plt.show()
```
